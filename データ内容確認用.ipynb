# データについて気が付いたこと
## customers
 customers_idがprimary keyとして使える。一意制約とnot nullを満たす。ordersにも登場.
 customer_unique_idがnot null だが uniqueではなさそう。customer_idとの違いが何か不明。
## geolocation
primary keyとして使えるものがわからない.
全体としても重要な情報は含まれていなさそう。
## order_items
order_id、product_idが外部キーそして使えそう。登場するテーブルorder_id：（order_payments、order_reviews、orders）、product_id：（product）
order_idに重複がある。=　一注文に複数商品が含まれている。
## order_payments 
order_idが外部キーとして使えそう。それ以外は支払方法の情報
order_idに重複あり=一注文に複数の支払い方法が含まれている。
## order_reviews
order_idが外部キーになりそう。それ以外はレビューの詳細
評価は1から5までの五段階評価。
order_id,review_idともに重複あり。=一つの注文に対して複数回レビューしている人がいるかも。
## orders
order_idが外部キーになりそう。それ以外は注文のタイミングや商品到着のタイミングなど注文の詳細
order_id,customer_idともに重複、欠損無し。=一つの注文につき一つのidが付いている。
## products
product_idが外部キーになりそう。
product_idに重複、欠損無し。→一つの商品に対して一つのidが降られている。
その他は送品の詳細情報。
## sellers 
seller_stateがcustomersのcustomer_stateと対応しているかも。確信は無し。
order_itemsにseller_idが登場。意味は今のところ不明。
